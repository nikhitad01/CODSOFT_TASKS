```python
# ============================================================
# Import Required Libraries
# ============================================================

import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
```

```python
# ============================================================
# Load Dataset
# ============================================================

DATA_PATH = r"C:\Users\Ramesh Darshanala\AdIntel\data\raw\advertising.csv"

df = pd.read_csv(DATA_PATH)

print("Dataset loaded successfully.")
```

```python
# ============================================================
# Statistical Summary
# ============================================================

display(df.describe())
```

```python
# ============================================================
# Missing Value Analysis
# ============================================================

missing_values = pd.DataFrame({
    "Missing Values": df.isnull().sum(),
    "Percentage": (df.isnull().sum()/len(df))*100
})

display(missing_values)
```

```python
# ============================================================
# Duplicate Analysis
# ============================================================

duplicates = df.duplicated().sum()

print(f"Duplicate Rows : {duplicates}")
```

```python
# ============================================================
# Column Names
# ============================================================

print("Columns")

for column in df.columns:
    print(column)
```

```python
# ============================================================
# Data Types
# ============================================================

display(df.dtypes)
```

```python
# ============================================================
# Unique Values
# ============================================================

unique_values = pd.DataFrame({
    "Unique Values": df.nunique()
})

display(unique_values)
```

```python
# ============================================================
# Remove Unnamed Column (if present)
# ============================================================

if "Unnamed: 0" in df.columns:
    df.drop(columns=["Unnamed: 0"], inplace=True)
    print("Unnamed column removed.")
else:
    print("No unnamed column found.")
```

```python
# ============================================================
# Dataset Profiling Summary
# ============================================================

profile_summary = pd.DataFrame({
    "Metric": [
        "Total Rows",
        "Total Columns",
        "Numerical Features",
        "Categorical Features",
        "Missing Values",
        "Duplicate Rows",
        "Target Variable"
    ],
    "Value": [
        df.shape[0],
        df.shape[1],
        len(df.select_dtypes(include=np.number).columns),
        len(df.select_dtypes(exclude=np.number).columns),
        df.isnull().sum().sum(),
        df.duplicated().sum(),
        "Sales"
    ]
})

display(profile_summary)
```
