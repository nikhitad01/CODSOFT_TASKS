```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)

print("Libraries imported successfully!")
```

```python
df = pd.read_csv("../data/imdb_eda.csv")

display(df.head())
print(df.shape)
```

```python
print(df.columns.tolist())
```

```python
required_cols = [
    "name", "year", "duration", "genre", "main_genre",
    "rating", "votes", "director", "actor_1", "actor_2", "actor_3"
]

df = df[required_cols].copy()

display(df.head())
```

```python
text_cols = ["name", "genre", "main_genre", "director", "actor_1", "actor_2", "actor_3"]

for col in text_cols:
    df[col] = df[col].astype(str).str.strip()
    df[col] = df[col].replace("nan", np.nan)

df["rating"] = pd.to_numeric(df["rating"], errors="coerce")
df["year"] = pd.to_numeric(df["year"], errors="coerce")
df["duration"] = pd.to_numeric(df["duration"], errors="coerce")
df["votes"] = pd.to_numeric(df["votes"], errors="coerce")

display(df.head())
```

```python
df = df.dropna(subset=["rating"])

print("Shape after removing rows without rating:", df.shape)
```

```python
df["duration_missing"] = df["duration"].isnull().astype(int)
df["votes_missing"] = df["votes"].isnull().astype(int)
df["year_missing"] = df["year"].isnull().astype(int)
df["director_missing"] = df["director"].isnull().astype(int)
df["actor_1_missing"] = df["actor_1"].isnull().astype(int)

display(df.head())
```

```python
df["duration"] = df["duration"].fillna(df["duration"].median())
df["votes"] = df["votes"].fillna(df["votes"].median())
df["year"] = df["year"].fillna(df["year"].median())

print(df[["duration", "votes", "year"]].isnull().sum())
```

```python
categorical_cols = ["genre", "main_genre", "director", "actor_1", "actor_2", "actor_3"]

for col in categorical_cols:
    df[col] = df[col].fillna("Unknown")

display(df.head())
```

```python
CURRENT_YEAR = 2026

df["movie_age"] = CURRENT_YEAR - df["year"]

display(df[["name", "year", "movie_age"]].head())
```

```python
df["genre_count"] = df["genre"].apply(lambda x: len(str(x).split(",")) if x != "Unknown" else 0)

display(df[["name", "genre", "genre_count"]].head())
```

```python
def runtime_category(minutes):
    if minutes < 90:
        return "Short"
    elif minutes <= 150:
        return "Medium"
    else:
        return "Long"

df["runtime_category"] = df["duration"].apply(runtime_category)

display(df[["name", "duration", "runtime_category"]].head())
```

```python
df["log_votes"] = np.log1p(df["votes"])

display(df[["name", "votes", "log_votes"]].head())
```

```python
vote_75 = df["votes"].quantile(0.75)
vote_90 = df["votes"].quantile(0.90)

df["popular_movie"] = (df["votes"] >= vote_75).astype(int)
df["highly_popular_movie"] = (df["votes"] >= vote_90).astype(int)
df["vote_percentile"] = df["votes"].rank(pct=True)

display(df[["name", "votes", "popular_movie", "highly_popular_movie", "vote_percentile"]].head())
```

```python
director_experience = df["director"].value_counts()

df["director_experience"] = df["director"].map(director_experience)

display(df[["name", "director", "director_experience"]].head())
```

```python
director_avg_rating = df.groupby("director")["rating"].mean()

df["director_avg_rating"] = df["director"].map(director_avg_rating)

display(df[["name", "director", "director_avg_rating"]].head())
```

```python
actor_experience = df["actor_1"].value_counts()

df["lead_actor_experience"] = df["actor_1"].map(actor_experience)

display(df[["name", "actor_1", "lead_actor_experience"]].head())
```

```python
actor_avg_rating = df.groupby("actor_1")["rating"].mean()

df["lead_actor_avg_rating"] = df["actor_1"].map(actor_avg_rating)

display(df[["name", "actor_1", "lead_actor_avg_rating"]].head())
```

```python
genre_frequency = df["main_genre"].value_counts(normalize=True)

df["genre_frequency"] = df["main_genre"].map(genre_frequency)

display(df[["name", "main_genre", "genre_frequency"]].head())
```

```python
runtime_dummies = pd.get_dummies(df["runtime_category"], prefix="runtime", drop_first=False)

df = pd.concat([df, runtime_dummies], axis=1)

display(df.head())
```

```python
genre_dummies = df["genre"].str.get_dummies(sep=",")
genre_dummies.columns = genre_dummies.columns.str.strip().str.lower().str.replace(" ", "_")
genre_dummies = genre_dummies.add_prefix("genre_")

df = pd.concat([df, genre_dummies], axis=1)

display(df.head())
```

```python
print("Final shape:", df.shape)

display(df.head())

display(df.info())
```

```python
missing_after = df.isnull().sum().sort_values(ascending=False)

display(missing_after[missing_after > 0])
```

```python
df.to_csv("../data/imdb_features.csv", index=False)

print("Feature engineered dataset saved successfully!")
print("File: ../data/imdb_features.csv")
```
